## $$ \text{STEP 3: CANARY CHANT} $$

After having understood the assets that the strategy considers, we define one of the most important functions of the strategy: $\fbox{13612U}$ Momentum Score

The strategy relies on the $\textcolor{yellow}{Canary}$ Universe to decide to which Universe we should be, either $\textcolor{green}{Defensive}$ Universe or $\textcolor{red}{Offensive}$ Universe.

The $\fbox{13612U}$ Momentum Score is defined as follows:

$$ 13612U_{ETF}() = (\underbrace{\frac{p_{0}-p_{1}}{p_{1}}}_{\text{1-Month Return}}) + (\underbrace{\frac{p_{0}-p_{3}}{p_{3}}}_{\text{3-Months Return}}) + (\underbrace{\frac{p_{0}-p_{6}}{p_{6}}}_{\text{6-Months Return}}) + (\underbrace{\frac{p_{0}-p_{12}}{p_{12}}}_{\text{12-Months Return}}) $$

where $p_{0}$ is the price of the ETF today, $p_{1}$ is the price of the ETF exactly 1-Month ago, $p_{3}$ is the price of the ETF 3-Months ago and as expected $p_{12}$ is the price of the ETF exactly 12-Months ago.

The uniform momentum score is a measure used to quantify the recent performance trend of an asset by equally weighing its returns over specified time periods. Unlike weighted momentum scores, which assign different weights to different periods to emphasize the importance of certain time frames over others, the uniform momentum score treats each time period the same.


How the $\textcolor{yellow}{Canary}$ Universe works?

1. If [$13612U_{TIP}$] > 0 then it suggests a trend of positive performance $\longrightarrow$ we switch to the $\textcolor{red}{Offensive}$ Universe.

2. If [$13612U_{TIP}$] < 0 then it suggests a downtrend $\longrightarrow$ we switch to the $\textcolor{green}{Defensive}$ Universe.


## $$ \text{STEP 4: NEW ASSETS SELECTION} $$

To explain the HAA (Hybrid Asset Allocation) strategy clearly, let's break it down into steps and illustrate with hypothetical cases. This strategy involves a mix of $\textcolor{red}{Offensive}$, $\textcolor{green}{Defensive}$, and $\textcolor{yellow}{Canary}$ assets to navigate various market conditions. Here's how it works:

1. $\fbox{Calculate Momentum}$

    For each asset in the offensive, defensive, and canary universes, calculate the unweighted average total return over the past 1, 3, 6, and 12 months. This is known as the 13612U momentum score.

2. $\fbox{Rank Assets}$

    Rank all assets within their respective universes (offensive, defensive, canary) based on their momentum scores.

3. $\fbox{Election Criteria}$

    $\textcolor{green}{Defensive}$ Strategy (When Canary Asset's Momentum is Non-Positive):
    
    If the $\textcolor{yellow}{Canary}$ asset (TIP) has a non-positive momentum [**13612U <= 0**], it signals potential market downturns or instability. In this case, allocate the portfolio to the best defensive "cash" asset (BIL or IEF) based on their momentum scores.
  
    $\textcolor{red}{Offensive}$ Strategy (When Canary Asset's Momentum is Positive):
    
    If the $\textcolor{yellow}{Canary}$ asset (TIP) has a positive momentum [**13612U > 0**], it signals favorable market conditions. Allocate 1/4 of the portfolio to each of the best Top 4 half of the risky assets (equally weighted). However, replace any of those Top 4 assets with the best defensive "cash" asset if it has non-positive momentum.

### $\fbox{Cases}$

#### $\fbox{Case A: Canary Asset Shows Negative Momentum}$

**Situation**: The $\textcolor{yellow}{Canary}$ asset TIP ETF has a negative 13612U momentum score at the end of the month.

**Action**: The entire portfolio is allocated to the $\textcolor{green}{Defensive}$ Universe, either BIL or IEF, depending on which has a better momentum score.

#### $\fbox{Case B: Canary Asset Shows Positive Momentum, but Some Offensive Assets Have Negative Momentum}$

**Situation**: The $\textcolor{yellow}{Canary}$ asset TIP ETF has a positive 13612U momentum score, indicating a favorable market. However, among the top offensive assets selected, a few have non-positive momentum.

**Action**: Allocate equally among the top half of the $\textcolor{red}{Offensive}$ assets with positive momentum. For any asset initially selected but showing non-positive momentum, replace it/them with the $\textcolor{green}{Defensive}$  "cash" asset (BIL or IEF) that has the best momentum score.

#### $\fbox{Case C: All Selected Offensive Assets Have Positive Momentum}$

**Situation**: The $\textcolor{yellow}{Canary}$ asset TIP shows positive momentum, and all top offensive assets also have positive momentum.

**Action**: Equally allocate the portfolio among the top half of these $\textcolor{red}{Offensive}$ assets without needing to replace any with defensive assets.

## $$ \text{STEP 5: REBALANCING \& HOLDING} $$

Finally, make sure to follow the last steps:

1. Equally-weight to all assets selected at the close:

When multiple assets are selected for investment, an equal amount of capital is allocated to each asset. For example, if the strategy dictates investing in four assets, then each asset would receive one-fourth of the total capital available for investment.
"At the close" typically means that these adjustments are made at the closing price of the assets at the end of the trading day.
Rebalance the portfolio even if there isn’t a change in signal:

2. Rebalancing is the process of realigning the weightings of the assets in a portfolio. 

This ensures that the intended asset allocation is maintained over time.
This particular guideline states that the rebalancing should occur every month, even if the strategy's indicators ("signals") have not changed since the last rebalancing. This could mean buying or selling assets to maintain the equal weighting, due to changes in asset values over the month.

3. Hold all positions until the end of the following month:

Once the portfolio is set at the close of the trading day (usually on the last day of the month), all positions are held without any further buying or selling until the last day of the next month.
This rule ensures that the strategy is not reacting to market movements during the month, which could incur additional transaction costs and potentially destabilize the strategic allocation.

# Python Implementation

## Step 1: Retrieval of Data

### Yahoo Finance Data Retrieval

```python
from datetime import date
import datetime
import yfinance as yf
import pandas as pd

canary_tickers = ["TIP"]
offensive_tickers = ["SPY","IWM","VEA","VWO","DBC","VNQ","TLT"]
defensive_tickers = ["BIL", "IEF"]
haa_tickers = ["SPY","IWM","VEA","VWO","DBC","VNQ","TLT","TIP", "BIL", "IEF"]


def fetch_data(tickers, start_date, end_date):
    data = yf.download(tickers, start=start_date, end=end_date)
    return data['Adj Close']


online = pd.DataFrame(fetch_data(haa_tickers, '2022-01-01', datetime.datetime.today()))

online
```

### Visualization of Data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

for ticker in offensive_tickers:
    plt.plot(online[ticker], label=ticker)

# Label the plot
plt.title('ETFs Adjusted Close Prices')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.legend()
plt.show()
```

### Data Manipulation:


#### Monthly Data

```python
online_monthly = online.resample("BM").last()
online_monthly.rename(index={online_monthly.index[-1]:datetime.datetime.today()},inplace=True)
online_monthly.index = online_monthly.index.strftime('%B %d, %Y')
online_monthly = online_monthly[:len(online_monthly)-1]
online_monthly
```

## Step 2: Momentum Calculation

### 13612U Momentum

```python
periods = [1,3,6,12]

# Function to compute the 13612U Momentum
def calculate_13612U_momentum(data_monthly, periods):
    uniform_momentum_score = sum([(data_monthly / data_monthly.shift(month) - 1) / 4 for month in periods]).dropna()
    return uniform_momentum_score*100
        
uniform_momentum_df = calculate_13612U_momentum(online_monthly, periods)
uniform_momentum_df
```

## Step 3: Asset Selection + Canary Chant

```python
canary_tickers = ["TIP"]
offensive_tickers = ["SPY","IWM","VEA","VWO","DBC","VNQ","TLT"]
defensive_tickers = ["BIL", "IEF"]

def canary_chant(uniform_momentum):

    # Universes of assets
    canary_asset = uniform_momentum[canary_tickers]
    offensive_assets = uniform_momentum[offensive_tickers]
    defensive_assets = uniform_momentum[defensive_tickers]
    
    # Access to the last row of Canary Assets 
    canary_assets = uniform_momentum[canary_tickers].iloc[-1]
    canary_assets = pd.DataFrame(canary_assets.map(lambda x: True if x > 0 else False)).transpose()
    
    # Access to Top Defensive Assets (Tickers and Values)
    top_defensive_assets = defensive_assets.iloc[-1].idxmax()
    top_defensive_assets_values = defensive_assets[top_defensive_assets].iloc[-1]
    
    # Access to Top Offensive Assets
    top_offensive_assets = offensive_assets.iloc[-1].nlargest(4).index.tolist()
    top_offensive_assets_values = offensive_assets[top_offensive_assets].iloc[-1].to_list()
    
    
    # Offensive assets to defensive positions
    off_to_def = [(asset, asset_value > 0) for asset, asset_value in zip(top_offensive_assets, top_offensive_assets_values)]

    # Cases to consider
    if all(canary_assets):
        print(f"In {canary_assets.index[0]} we switch to Offensive Universe.\n")
        print(f"    Offensive Assets (Top 4/8): {top_offensive_assets}")
        print(f"    Who remains in Offensive Assets? {off_to_def}\n")

    else:
        print(f"In {canary_assets.index[0]} we switch to Defensive Universe.\n")
        print(f"    Defensive Assets (Top 1/2): {top_defensive_assets}")

canary_chant(uniform_momentum_df)
```
